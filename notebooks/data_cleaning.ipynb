```python
import pandas as pd

df = pd.read_csv("../data/raw/layoffs.csv")

df.head()
```

```python
df.isnull().sum()
```

```python
df['industry'] = df['industry'].fillna('Unknown')
```

```python
df = df.dropna(subset=['total_laid_off', 'percentage_laid_off'], how='all')
```

```python
df['industry'] = df['industry'].str.strip().str.title()
df['country'] = df['country'].str.strip().str.title()
```

```python
df['country'] = df['country'].replace({
    'United States.': 'United States',
    'Usa': 'United States'
})
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
```

```python
df['total_laid_off'] = pd.to_numeric(df['total_laid_off'], errors='coerce')
df['funds_raised_millions'] = pd.to_numeric(df['funds_raised_millions'], errors='coerce')
```

```python
df = df.drop_duplicates()
```

```python
df.to_csv("../data/processed/layoffs_cleaned.csv", index=False)
```
